# Visualizing Earnings Based On College Majors

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads=pd.read_csv('C:/Users/dasarr/Downloads/DDDD/DDDD/data-master/recent-grads.csv')
recent_grads.head()
```

```python
recent_grads.count()
```

```python
recent_grads=recent_grads.dropna()
```

```python
recent_grads.count()
```

```python
recent_grads.plot(x='Sample_size',y='Median',kind='scatter',title='Sample_size vs Median',color='red')
recent_grads.plot(x='Sample_size',y='Unemployment_rate',kind='scatter',title='Sample_size vs Unemployment_rate',color='blue')
```

The median salary ranges from 20000 to 70000 for majority of the smaple and very dense distributionn for the salary ranges of 100000
The unemployment rate is in the range of 0.020 to 0.120 for majority of sample

```python
recent_grads.plot(x='Men',y='Median',kind='scatter',title='Men vs Median',color='red')
recent_grads.plot(x='Women',y='Median',kind='scatter',title='Women vs Median',color='blue')
recent_grads.plot(x='Full_time',y='Median',kind='scatter',title='Full_time vs Median',color='green')

```

Comparing the plots for Men And Women we can say that the women median salary is higher than men.
